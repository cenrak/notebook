```python
%matplotlib inline
from matplotlib import pyplot as plt
male_age = range(20, 71)
fixed_theory = lambda x: 0.8 * x
dynamic_theory = lambda x: ( 1 - (750 / x)/100) * x
dynamic_theory2 = lambda x: ( 1 - (1500 / x**2) / 100) * x
age = lambda x, y: x/100 * y
```

```python
fixed = map(fixed_theory, male_age)
dynamic = map(dynamic_theory, male_age)
dynamic2 = map(dynamic_theory2, male_age)
```

```python
plt.plot(list(male_age), list(fixed))
plt.scatter(list(male_age), list(dynamic))
plt.scatter(list(male_age), list(dynamic2))
plt.legend(['fixed', 'dynamic', 'dynamic2'])
```
